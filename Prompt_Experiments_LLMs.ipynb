**Text Generation (GPT-2):**

```python
pipe = pipeline("text-generation", model="gpt2")
#pipeline("text-generation", model="distilbert/distilgpt2")

prompt1 = "My name is Ran and  "
prompt2 = "My name is Ran and in 3 points this is what  "

output1=pipe(prompt1)
output2=pipe(prompt2,max_new_tokens=50,top_k=50)

print("Prompt1:\n",output1[0]["generated_text"])
print("Prompt2:\n",output2[0]["generated_text"])
```

-----------------
1. Prompt1:
-----------------
fot
* No special request was given except for generating a new text for this prompt.
* The output shows that the text was not so clear and contained long sentences.

-----------------
2. Prompt2:
-----------------

* The model was requested to give 3 points of something after the same prompts , we limit the length of the output and used top_k to choose the next token from the 50 most probability.
* The output shows that the model could not finish the first sentence clearly , but he still gave some 3 points of something. The text become personlly and humen.
* The conclusion is that the model can effectively control length, but it can't control on the structure.


**Summarization (Bert):**

```python
pipe_sum = pipeline("summarization", model="facebook/bart-large-cnn")

text = """
The concept of animal rights is highly topical in animal legal scholarship.
There is much debate on whether or not animals can and do hold rights,
although there does seem to be consensus that animals currently do not have
such legal rights that would grant them any foundational protection in the
way that human rights do for humans.
This article challenges that assumption through a discussion on the legal
rights of wild animals under the European Union (EU) Habitats Directive.
The article establishes a theoretical framework on animal rights and analyses
the Habitats Directive according to the rights framework.
It subsequently argues that preliminary fundamental legal rights for certain
wild animals can be found in EU conservation law.
However, through a discussion on the scope and objective of the Habitats
Directive, the article finds that the legal rights that are found are not
grounded in the interests of the animals themselves.
Instead, the preliminary fundamental animal rights are a by-product of the
anthropocentric value of biodiversity and do not support a rights-based
approach to animal rights.
The article concludes with a discussion on the interpretation that follows
from these findings and how an instrument such as the Habitats Directive
could implement rights-based wild animal rights.
"""
output1=pipe_sum(text)
output2=pipe_sum("summerize the follow for a childe: "+text, max_length=56)
print("Prompt1:\n",output1[0]["summary_text"])
print("Prompt2:\n",output2[0]["summary_text"])
```

-----------------
1. Prompt1:
-----------------

* No special request was given except for summarizing the article.
* The output shows that he did a clear summary that explained the main concepts of the article.

-----------------
2. Prompt2:
-----------------

* The model was requested to summarize the article for a child with a max length of 56 characters.
* The output shows that the model focused on the length request, but could not change it to a child understanding: he is still using complex vocabulary.
* The conclusion is that giving instructions to summarize a text for a specific audience is not something this model can do. The model can effectively control length and condense content, but it can't modify style or simplify language for a specific audience.
